# RECURSION (Özyineleme)

Recursion; önemsiz bir şekilde çözülebilecek kadar küçük bir soruna ulaşana kadar, bir sorunu daha küçük alt problemlere bölmeyi içeren bir problem çözme yöntemidir. Recursion genellikle, kendisini çağıran bir işlevi içerir. Yüzeysel olarak bakıldığında pek bir şey gibi görünmese de, programlaması çok zor olabilecek sorunlara zarif çözümler yazmamıza olanak tanır. 

**Calculating the Sum of a List of Numbers (Bir listedeki sayıların toplamı) :**

Aşağıdaki kod bloklarında ilk önce recursion kullanmadan bu problemi çözeceğiz. Daha sonra recursion kullanarak ile çözeceğiz. 

```python
def listsum(numList):
    """Recursion kullanmadan listedeki sayıları toplayan fonksiyon"""
    theSum = 0
    for i in numList:
        theSum += i
    return theSum

print(listsum([1,3,5,7,9]))
```

Bu fikri bir Python programına şu şekilde dönüştürebiliriz: NumList listesinin toplamı, listenin ilk öğesinin (numList[0]) ve listenin geri kalanındaki sayıların toplamıdır (numList[1:]) diyebiliriz. İşlevsel bir biçimde şu şekilde ifade edebiliriz:



```python
def listsum(numList):
    """Recursion ile listedeki sayıların toplamını bulan fonksiyon"""
    if len(numList) == 1:
        return numList[0]
    else:
        return numList[0] + listsum(numList[1:])
    
print(listsum([1,3,5,7,9]))
```

+ Else bloğu altında recursion (özyineleme) vardır. # The Three Laws Of Recursion (Özyinelemenin 3 Yasası) 

Tüm özyinelemeli algoritmalar şu 3 yasaya uymak zorundadır :

+ 1-) Özyinelemeli bir algoritmanın, bir temel durumu olmalıdır.
+ 2-) Özyinelemeli bir algoritma, durumunu değiştirmeli ve temel duruma doğru hareket etmelidir.
+ 3-) Özyinelemeli bir algoritma, kendini yinelemeli olarak çağırmalıdır.

**1-) Özyinelemeli bir algoritmanın, bir temel durumu olmalıdır.**

**Temel durum** olarak ifade ettiğimiz durum, algoritmanın yinelemeyi durdurmasına izin veren koşuldur. **Temel durum**, problemin doğrudan çözülecek kadar küçük bir hale gelmiş halidir. **listsum algoritmasında (yukarıdaki örnek) temel durum**, 1 uzunluğunda bir listedir.

**2-) Özyinelemeli bir algoritma, durumunu değiştirmeli ve temel duruma doğru hareket etmelidir.**

Bu yasaya uymak için, algoritmayı temel duruma doğru hareket ettiren bir **durum değişikliği** ayarlamalıyız. **Durum değişikliği**, algoritmanın kullandığı bazı verilerin değiştirildiği anlamına gelir. Genellikle sorunumuzu temsil eden verilen bir şekilde küçülür. **listsum algoritmasında (yukarıdaki örnek)** birincil veri yapımız bir listeydi. Bu listeyi temel duruma doğru, doğal bir ilerleme ile kısalttık ve temel duruma, uzunluğu 1 olan bir liste haline getirdik.   

**3-) Özyinelemeli bir algoritma, kendini yinelemeli olarak çağırmalıdır.**

Nihai yasa, algoritmanın kendisini çağırması gerektiğidir. Özyinelemenin tam tanımı budur.

Aşağıdaki örnekte, girilen sayının faktöriyelini bulan fonksiyonu recursion ile yazalım. 

```python
def fact(n):
    """Girilen sayının faktöriyelini recursion ile bulan fonksiyon"""
    if n <= 1: 
        return 1
    else: 
        return n * fact(n-1)

print(fact(5))
print(fact(6))
print(fact(7))
```

+ **if n <= 1** bloğunun altındaki kod satırı, özyinelemeli algoritmanın **temel durumudur**. Eğer girilen sayı 0 veya 1 ise ya da recursion sayesinde temel duruma ulaşıldığında 1 döndürülecek. **else bloğunun** altındaki kod satırında, girilen sayı 0 veya 1 den büyük iken bir özyineleme gerçekleşecek ve **n <= 1** olana kadar fonksiyon kendi içinde çağırılacaktır. Yani **else bloğu** altında algoritmanın durumu değişecek ve temel duruma doğru hareket edecektir.

Aşağıdaki örnekte, bir tam sayıyı herhangi bir tabanda dizeye dönüştüren fonksiyonu recursion olacak şekilde yazalım. 

```python
def toStr(n, base):
    convertString = "0123456789ABCDEF"
    if n < base:
        return convertString[n]
    else:
        return toStr(n//base, base) + convertString[n%base]
        
print(toStr(1453, 16))
```

+ Yukarıdaki örnekte, 1453 sayısını 16'lık tabanda bir dizeye dönüştürmeye çalışıyoruz. Dönüştüreceğimiz dizi **converString** adlı değişkendeki harflerden oluşacak. Bir sayıyı, başka bir sayılık tabanda yazma işlemi şu şekilde yapılır: Dönüşecek olan sayı, dönüştürülecek taban sayısına bölünür ve bölüm sonucu elde edilen sayı en yakın tam sayıya yuvarlanır. Bu işlem Python'da **floor ( // )** operatörü ile yapılır. Biz bu fonksiyonda, bölümden elde edilen sonucu ayrıca bir metine karşılık getirmek istiyoruz. İf bloğu, eğer dönüştürülmek istenen sayı, dönüştürülecek olan taban sayısından küçükse direkt olarak **convertString** değişkeninde karşılık geldiği indexin metnine eşit olduğunu ifade eder. Bu bizim temel durumumuzdur ve else bloğunda, dönüştürülecek olan sayıyı bu temel duruma indirgemeye çalışacağız. Else bloğunda ise şu işlem gerçekleşmektedir : 



Aşağıdaki kod bloğunda, girilen metnin palindrom olup olmadığını bulan bir fonksiyonu recursion kullanarak yazalım.

```python
def palindrome(string):
    """Girilen metnin palindrom olup olmadığını recursion ile bulan fonksiyon"""
    if len(string) <= 1: #Temel durumumuz
        return True
    else: #Girilen metni, temel duruma indirgemek için yaptığımız işlem.
        return string[0] == string[-1] and palindrome(string[1:-1])
    
print(palindrome("kayak"))
```

# Dynamic Programming (Dinamik Programlama)

Dinamik programlama, optimizasyon problemleri için bir stratejidir.

Örnek olarak bir optimizasyon probleminin klasik bir örneği, en az parayı kullanarak değişiklik yapmayı içerir. Bir satış makinesi üreticisi için programcı olduğunuzu varsayalım. Şirketiniz, her işlem için mümkün olan en az madeni para bozdurarak çabayı kolaylaştırmak istiyor. Bir müşterinin bir dolarlık banknot koyduğunu ve 37 sente bir ürün satın aldığını varsayalım. Değişiklik yapmak için kullanabileceğiniz en az jeton sayısı nedir? Cevap altı madeni para: iki çeyrek, bir kuruş ve üç peni. Altı madeni paranın cevabına nasıl ulaştık? Cephaneliğimizdeki en büyük jetonla (çeyrek) başlıyoruz ve mümkün olduğu kadar çok kullanıyoruz, ardından bir sonraki en düşük jeton değerine gidiyoruz ve mümkün olduğu kadar çok kullanıyoruz. Bu ilk yaklaşıma **greedy method (açgözlü yöntem)** denir, çünkü **sorunun olabildiğince büyük bir parçasını hemen çözmeye çalışırız.**

ABD madeni paraları kullandığımızda açgözlü yöntem gayet iyi işliyor, ancak şirketinizin otomatlarını Aşağı Elbonia'da kurmaya karar verdiğini varsayalım, burada her zamanki 1, 5, 10 ve 25 sentlik madeni paralara ek olarak 21 sentlik bir madeni para da var. Bu örnekte **greedy method (açgözlü yöntemimiz)**, 63 sentlik değişim için en uygun çözümü bulamıyor. 21 sentlik madeni paranın eklenmesiyle, açgözlü yöntem yine de çözümü altı madeni para olarak bulacaktır. Bununla birlikte, en uygun cevap üç adet 21 senttir.

Soruna en uygun yanıtı bulacağımızdan emin olabileceğimiz bir yönteme bakalım ve bunu recursion ile yapalım. İlk olarak **temel durumu** belirleyelim. Madeni paralarımızdan birinin değeri kadar bozdurmaya çalışıyorsak cevap kolay, tek madeni para. Yani 1, 5, 10, 25 ve 21 değerinde bir madeni para bozuyorsak direkt olarak bu paralardan 1 tanesini vereceğimiz için cevap 1 olacak. 

Eğer bozdurulacak para eşleşmezse şu şekilde bir yöntem izleyebiliriz:

 

Aşağıdaki kod bloğunda, yukardaki numCoins olarak örnek gösterdiğimiz yöntemi bir fonksiyon olarak kodlayalım. 

```python
def recMC(coinValueList, change):
    minCoins = change
    if change in coinValueList:
        return 1
    else:
        for i in [c for c in coinValueList if c <= change]:
            numCoins = 1 + recMC(coinValueList, change - i)
            if numCoins < minCoins:
                minCoins = numCoins
                
    return minCoins

print(recMC([1,5,10,25],63))
```

+ Üçüncü satırda temel durumu kontrol ediyoruz; yani madeni paralarımızdan birinin tam miktarını bozdurmaya çalışıyoruz. Bozuk para miktarına eşit bir madeni paramız yoksa, yapmaya çalıştığımız bozuk para miktarından daha az olan her bir farklı madeni para değeri için yinelemeli aramalar yaparız. 6. Bu, algoritmanın her adımında, daha küçük bir miktar için değişiklik yapmak için gereken minimum jeton sayısını zaten bildiğimizi garanti eder.

11 sentlik bozdurmada kullanılacak olan minimum madeni para tablosu: 



11 sentlik bozdurmada kullanılacak minimum madeni para tablosunu nasıl dolduracağımıza bakalım. Şekil 4 işlemi göstermektedir. Bir sentle başlıyoruz. Mümkün olan tek çözüm bir madeni paradır (bir kuruş). Sonraki satır, bir sent ve iki sent için minimum değeri gösterir. Yine, tek çözüm iki kuruş. Beşinci sıra, işlerin ilginçleştiği yerdir. Şimdi düşünmemiz gereken iki seçenek var, beş kuruş ya da bir beş sent. Hangisinin en iyi olduğuna nasıl karar veririz? Tabloya bakarız ve dört sentlik bozuk para yapmak için gereken madeni para sayısının dört, artı beş için bir kuruş daha beş madeni paraya eşit olduğunu görürüz. Veya beş sentin 1 madeni paraya eşit olması için sıfır sent artı bir nikele daha bakabiliriz. Bir ve beşin minimumu bir olduğu için tabloda 1 saklıyoruz. Tablonun sonuna tekrar hızlıca ilerleyin ve 11 senti düşünün. Şekil 5, dikkate almamız gereken üç seçeneği göstermektedir:



Aşağıdaki kod bloğunda, para bozdurma algoritmasını, dinamik programlama ile çözelim. dpMakeChange üç parametre alır: geçerli madeni para değerleri listesi, yapmak istediğimiz değişiklik miktarı ve her değeri oluşturmak için gereken minimum madeni para sayısının listesi. İşlev bittiğinde minCoins, 0'dan değişim değerine kadar tüm değerler için çözümü içerecektir.

```python
def dpMakeChange(coinValueList, change, minCoins):
    for cents in range(change + 1):
        coinCount = cents
        for j in [c for c in coinValueList if c <= cents]:
            if minCoins[cents - j] + 1 < coinCount:
                coinCount = minCoins[cents - j] + 1
        minCoins[cents] = coinCount
        
    return minCoins[change]
```

+ Bu soruna özyinelemeli bir çözümle başlamış olsak da, dpMakeChange'in özyinelemeli bir işlev olmadığına dikkat edin. Bir soruna özyinelemeli bir çözüm yazabilmenizin, onun en iyi veya en verimli çözüm olduğu anlamına gelmediğini anlamak önemlidir. Bu işlevdeki işin büyük kısmı, 4. satırda başlayan döngü tarafından yapılır. Bu döngüde, sentlerle belirtilen miktarda değişiklik yapmak için olası tüm madeni paraları kullanmayı düşünüyoruz. Yukarıdaki 11 cent örneğinde yaptığımız gibi minimum değeri hatırlıyoruz ve minCoins listemizde saklıyoruz.

Para bozdurma algoritmamız minimum para sayısını bulmakta iyi bir iş çıkarsa da kullandığımız paraları takip etmediğimiz için değişiklik yapmamıza yardımcı olmuyor. MinCoins tablosundaki her giriş için eklediğimiz son parayı hatırlayarak kullanılan paraları takip etmek için dpMakeChange'i kolayca genişletebiliriz. Eklenen son madeni parayı biliyorsak, o miktarı yapmak için eklediğimiz son parayı bize söyleyen tabloda önceki bir girişi bulmak için madeni paranın değerini kolayca çıkarabiliriz. Başa gelene kadar tablonun izini sürmeye devam edebiliriz. ActiveCode 2, kullanılan madeni paraları takip etmek için değiştirilmiş dpMakeChange algoritmasını ve kullanılan her madeni paranın değerini yazdırmak için tabloda geriye doğru yürüyen printCoins işlevini gösterir. Bu, Aşağı Elbonia'daki dostlarımız için sorunu çözen algoritmayı iş başında gösteriyor. Ana ekranın ilk iki satırı dönüştürülecek miktarı belirler ve kullanılan madeni paraların listesini oluşturur. Sonraki iki satır, sonuçları saklamamız gereken listeleri oluşturur. coinUsed, değişiklik yapmak için kullanılan paraların bir listesidir ve coinCount, listedeki konuma karşılık gelen miktar için değişiklik yapmak için kullanılan minimum para sayısıdır. Yazdırdığımız madeni paraların doğrudan CoinUsed dizisinden geldiğine dikkat edin. İlk çağrı için dizi pozisyonu 63'ten başlayıp 21'i yazdırıyoruz. Ardından 63 − 21 = 42'yi alıp listenin 42. elemanına bakıyoruz. Bir kez daha orada saklanan bir 21 buluyoruz. Son olarak, dizinin 21. öğesi de 21 içerir ve bize 21 sentlik üç parça verir. 

```python
def dpMakeChange(coinValueList,change,minCoins,coinsUsed):
   for cents in range(change+1):
      coinCount = cents
      newCoin = 1
      for j in [c for c in coinValueList if c <= cents]:
            if minCoins[cents-j] + 1 < coinCount:
               coinCount = minCoins[cents-j]+1
               newCoin = j
      minCoins[cents] = coinCount
      coinsUsed[cents] = newCoin
   return minCoins[change]

def printCoins(coinsUsed,change):
   coin = change
   while coin > 0:
      thisCoin = coinsUsed[coin]
      print(thisCoin)
      coin = coin - thisCoin

def main():
    amnt = 63
    clist = [1,5,10,21,25]
    coinsUsed = [0]*(amnt+1)
    coinCount = [0]*(amnt+1)

    print("Making change for",amnt,"requires")
    print(dpMakeChange(clist,amnt,coinCount,coinsUsed),"coins")
    print("They are:")
    printCoins(coinsUsed,amnt)
    print("The used list is as follows:")
    print(coinsUsed)

main()
```
